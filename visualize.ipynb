```python
import math
import matplotlib.pyplot as plt

data = open('result.txt').read().split('\n')

level = [float(x) for x in data[0].split(' ') if x != '']
varZ =  [float(x) for x in data[1].split(' ') if x != '']
varP =  [float(x) for x in data[2].split(' ') if x != '']
aveZ =  [float(x) for x in data[3].split(' ') if x != '']
aveP =  [float(x) for x in data[4].split(' ') if x != '']
Kurtosis =  [float(x) for x in data[5].split(' ') if x != '']
eps =  [float(x) for x in data[6].split(' ') if x != '']
num = len(eps)
idx = 7
N = []
for i in range(num):
    N.append([float(x) for x in data[idx].split(' ') if x != ''])
    idx += 1

mlmc_cost = [float(x) for x in data[idx].split(' ') if x != '']
std_cost = [float(x) for x in data[idx + 1].split(' ') if x != '']
    
plt.figure(figsize = (18, 12))
sz = len(varZ)

# var
plt.subplot(2, 3, 1)
plt.xlabel('level l')
plt.ylabel('log2 variance')
plt.plot(level[:sz], varZ, label = 'Z')
plt.plot(level[:sz], varP, label = 'P')
plt.legend()

# ave
plt.subplot(2, 3, 2)
plt.xlabel('level l')
plt.ylabel('log2 |mean|')
plt.plot(level[:sz], aveZ, label = 'Z')
plt.plot(level[:sz], aveP, label = 'P')
plt.legend()

# kurtosis
plt.subplot(2, 3, 3)
plt.xlabel('level l')
plt.ylabel('Kurtosis')
plt.plot(level[:sz], Kurtosis)

# N
plt.subplot(2, 3, 4)
plt.xlabel('level l')
plt.ylabel('N')
for i in range(num):
    sz = len(N[i])
    plt.plot(level[:sz], N[i], label = str(eps[i]))
plt.legend()

# Cost
plt.subplot(2, 3, 5)
plt.xlabel('accuracy')
plt.ylabel('^2 Cost')
eps = [math.log10(x) for x in eps]
plt.plot(eps, mlmc_cost, label = 'mlmc cost')
plt.plot(eps, std_cost, label = 'std cost')
plt.legend()
```
